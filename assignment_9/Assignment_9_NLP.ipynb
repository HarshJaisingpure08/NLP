```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
sentences = [
    "यह फिल्म बहुत अच्छी है",
    "यह खाना बहुत खराब है",
    "मुझे यह मोबाइल पसंद है"
]

for s in sentences:
    result = sentiment_pipeline(s)
    print("Sentence:", s)
    print("Sentiment:", result)
    print()
```

```python
def sentiment_label(text):
    result = sentiment_pipeline(text)[0]

    stars = int(result['label'][0])

    if stars >= 4:
        return "Positive"
    elif stars == 3:
        return "Neutral"
    else:
        return "Negative"


for s in sentences:
    print(s, "->", sentiment_label(s))
```

```python
marathi_sentences = [
    "हा चित्रपट खूप छान आहे",
    "हे अन्न खूप वाईट आहे"
]

for s in marathi_sentences:
    print("Sentence:", s)
    print("Sentiment:", sentiment_pipeline(s))
```
