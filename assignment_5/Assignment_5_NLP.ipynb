```python
import nltk
from nltk.corpus import wordnet

# Download WordNet data
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "The car is very fast and the driver is happy"
words = text.split()

print("Words in sentence:")
print(words)
```

```python
def get_synonyms(word):
    synonyms = set()

    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())

    return synonyms


for word in words:
    print(f"\nSynonyms for '{word}':")
    print(get_synonyms(word))
```

```python
def get_antonyms(word):
    antonyms = set()

    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

    return antonyms


for word in words:
    ant = get_antonyms(word)
    if ant:
        print(f"\nAntonyms for '{word}':")
        print(ant)
```

```python
def get_hypernyms(word):
    hypernyms = set()

    for syn in wordnet.synsets(word):
        for hyper in syn.hypernyms():
            hypernyms.add(hyper.name())

    return hypernyms


for word in words:
    print(f"\nHypernyms for '{word}':")
    print(get_hypernyms(word))
```
