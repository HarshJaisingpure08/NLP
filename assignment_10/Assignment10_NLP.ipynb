```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
from collections import defaultdict, Counter
```

```python
text = """
I want to eat pizza
I want to eat pasta
I want to drink coffee
I want to drink tea
I like machine learning
I like deep learning
Machine learning is powerful
Deep learning is part of artificial intelligence
Artificial intelligence is the future
"""
```

```python
class NGramAutocomplete:

    def __init__(self, n=3):
        self.n = n
        self.ngram_counts = defaultdict(Counter)

    def train(self, text):
        words = word_tokenize(text.lower())

        for i in range(len(words) - self.n + 1):
            context = tuple(words[i:i+self.n-1])
            next_word = words[i+self.n-1]
            self.ngram_counts[context][next_word] += 1

    def predict(self, text, top_k=3):
        words = word_tokenize(text.lower())

        if len(words) < self.n - 1:
            return []

        context = tuple(words[-(self.n-1):])

        if context in self.ngram_counts:
            predictions = self.ngram_counts[context].most_common(top_k)
            return [word for word, count in predictions]

        return []
```

```python
model = NGramAutocomplete(n=3)

model.train(text)

print("Model trained successfully!")
```

```python
model = NGramAutocomplete(n=3)

model.train(text)

print("Model trained successfully!")
```

```python
print("Input: I want")
print("Suggestions:", model.predict("I want"))

print("\nInput: I want to")
print("Suggestions:", model.predict("I want to"))

print("\nInput: machine learning")
print("Suggestions:", model.predict("machine learning"))
```

```python
while True:
    user_input = input("Type a sentence (or 'exit'): ")

    if user_input.lower() == "exit":
        break

    suggestions = model.predict(user_input)

    print("Auto-complete suggestions:", suggestions)
```
