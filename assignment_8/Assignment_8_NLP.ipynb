```python
import nltk
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
```

```python
import nltk

# Download all required datasets
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk.tokenize import word_tokenize

sentence = "I went to the bank to deposit money."

words = word_tokenize(sentence)

print(words)
```

```python
sense = lesk(words, "bank")

print("Selected Sense:", sense)
print("Definition:", sense.definition())
print("Example:", sense.examples())
```

```python
for syn in wordnet.synsets("bank"):
    print("Sense:", syn.name())
    print("Definition:", syn.definition())
    print()
```

```python
sentence = "The fisherman sat on the bank of the river."

words = word_tokenize(sentence)

sense = lesk(words, "bank")

print("Selected Sense:", sense)
print("Definition:", sense.definition())
```

```python
def disambiguate(sentence, target_word):
    words = word_tokenize(sentence)
    sense = lesk(words, target_word)

    if sense:
        print("Sentence:", sentence)
        print("Word:", target_word)
        print("Meaning:", sense.definition())
    else:
        print("No sense found")


disambiguate("I deposited cash at the bank", "bank")
disambiguate("The children played near the river bank", "bank")
```
